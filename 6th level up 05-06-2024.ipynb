```python
df.isnull().sum()
```

```python
df.columns
```

```python
new_columns = ['use', 'gen', 'House overall', 'Dishwasher', 'Furnace 1', 'Furnace 2',
       'Home office', 'Fridge', 'Wine cellar', 'Garage door', 'Kitchen 12',
       'Kitchen 14', 'Kitchen 38', 'Barn', 'Well', 'Microwave', 'Living room',
       'Solar', 'temperature','humidity', 'visibility', 'apparentTemperature',
       'pressure', 'windSpeed', 'cloudCover', 'windBearing', 'precipIntensity',
       'dewPoint', 'precipProbability']
```

```python
for j in new_columns:
    Q1=np.percentile(df[j],25,method='midpoint')
    Q2=np.percentile(df[j],50,method='midpoint')
    Q3=np.percentile(df[j],75,method='midpoint')
    IQR = Q3-Q1
    low_lim = Q1-1.5*IQR
    up_lim = Q3+1.5*IQR
    outliers = []
    for h in df[j]:
        if ((h < low_lim) | (h > up_lim)):
            outliers.append(h)
    df[j]=df[j].clip(lower=low_lim,upper=up_lim)
```

# MODELLING

# # invoking train test split

```python
y = df.summary
X = df.drop('summary',axis=True)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix, precision_score, recall_score, f1_score
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

# Invoke the best model i.e. DECISION TREE without SCALING:-

# DECISION TREES

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)
decision_tree_pred = decision_tree.predict(X_test)
decision_tree_accuracy = accuracy_score(y_test, decision_tree_pred)
decision_tree_precision = precision_score(y_test, decision_tree_pred, average='macro')
decision_tree_recall = recall_score(y_test, decision_tree_pred, average='macro')
decision_tree_f1 = f1_score(y_test, decision_tree_pred, average='macro')
print("Decision Tree Accuracy:", decision_tree_accuracy)
print("confusion_matrix:", confusion_matrix(y_test,decision_tree_pred))
print("precision_score:",decision_tree_precision)
print("recall_score:",decision_tree_recall)
print("f1_score:",decision_tree_f1)
```

# cross valiation using stratified KFold method for decision tree

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
import statistics as stat
```

```python
strat_validator = StratifiedKFold(n_splits=10)
```

```python
skr_cv_result2 = cross_val_score(decision_tree,X,y,cv=strat_validator)
skr_cv_result2
```
